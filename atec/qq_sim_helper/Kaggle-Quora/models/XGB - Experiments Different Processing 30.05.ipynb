```python
import pandas as pd
import numpy as np
import time
import gc
import matplotlib.pyplot as plt
import xgboost as xgb
import lightgbm as lgb
%matplotlib inline
import seaborn as sns
sns.set_style("dark")
import warnings
warnings.filterwarnings('ignore')

from models_utils_gbm import *
from models_utils_fe import *
```

```python
def get_train():
    feats_src = '/media/w/1c392724-ecf3-4615-8f3c-79368ec36380/DS Projects/Kaggle/Quora/data/features/lemmatized_fullclean/'
    
    eda = pd.read_csv(feats_src + 'train_eda_features.csv').iloc[:, -1]
    srk = pd.read_csv(feats_src + 'train_SRKgrams_features.csv')
    #owl = pd.read_csv(feats_src + 'train_the1owl_features.csv').iloc[:, 9:]
    whq = pd.read_csv(feats_src + 'train_whq_with_jaccard_feats.csv')
    
    df = pd.concat([eda, srk, whq], axis = 1)
    df = drop_duplicate_cols(df)
    df = df.astype('float32')
    print('Training data shape:', df.shape)
    return df

def get_test():
    feats_src = '/media/w/1c392724-ecf3-4615-8f3c-79368ec36380/DS Projects/Kaggle/Quora/data/features/lemmatized_fullclean/'
    
    eda = pd.read_csv(feats_src + 'test_eda_features.csv').iloc[:, -1]
    srk = pd.read_csv(feats_src + 'test_SRKgrams_features.csv')
    #owl = pd.read_csv(feats_src + 'test_the1owl_features.csv').iloc[:, 9:]
    whq = pd.read_csv(feats_src + 'test_whq_with_jaccard_feats.csv')
    
    df = pd.concat([eda, srk, whq], axis = 1)
    df = drop_duplicate_cols(df)
    df = df.astype('float32')
    print('testing data shape:', df.shape)
    return df
```

```python
feats_src = '/media/w/1c392724-ecf3-4615-8f3c-79368ec36380/DS Projects/Kaggle/Quora/data/features/uncleaned/'

X_train = get_train()
xgb_feats = pd.read_csv(feats_src + '/the_1owl/owl_train.csv')
y_train = xgb_feats[['is_duplicate']]

X_test = get_test()
```

```python
lgb_params = {
    'task' : 'train',
    'boosting_type' : 'gbdt',
    'objective' : 'binary',
    'metric' : {'binary_logloss'},
    'learning_rate' : 0.01,
    'feature_fraction' : 0.53,
    'bagging_fraction': 0.9,
    'bagging_freq': 100,
    'num_leaves' : 255,
    'max_depth': 8,
    'min_data_in_leaf': 28,
    'subsample': 0.81,
    'colsample_bytree': 0.4,
    'silent': 1,
    'random_state': 1337,
    'verbose': 1,
    'nthread': 9,
    }

xgb_params = {
    'seed': 1337,
    'colsample_bytree': 0.42,
    'silent': 1,
    'subsample': 0.85,
    'eta': 0.05,
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'max_depth': 8,
    'min_child_weight': 20,
    'nthread': 4,
    'tree_method': 'hist',
    }


xgb = True

if xgb:
    run_xgb(X_train, y_train, xgb_params, 'diff_preprocessing_experiments_lemmatClean')
else:
    run_lgb(X_train, y_train, lgb_params, 'diff_preprocessing_experiments_lemmatClean')
```

```python
gbm = xgb.Booster(model_file = 'saved_models/XGB/XGB_diff_preprocessing_experiments_lemmatClean_0.44618.txt')
dtrain = xgb.DMatrix(X_train, label = y_train)

mapper = {'f{0}'.format(i): v for i, v in enumerate(dtrain.feature_names)}
importance = {mapper[k]: v for k, v in gbm.get_fscore().items()}
importance = sorted(importance.items(), key=lambda x:x[1], reverse=True)[:15]

df_importance = pd.DataFrame(importance, columns=['feature', 'fscore'])
df_importance['fscore'] = df_importance['fscore'] / df_importance['fscore'].sum()

plt.figure()
df_importance.plot()
df_importance.plot(kind='barh', x='feature', y='fscore', legend=False, figsize=(10, 18))
plt.title('XGBoost Feature Importance')
plt.xlabel('relative importance')
```

```python
Xtr2 = X_train.loc[:, df_importance.feature]
Xtr2.to_pickle('Xtrain_lemmatClean_15bestFeats.pkl')

Xte2 = X_test.loc[:, df_importance.feature]
Xte2.to_pickle('Xtest_lemmatClean_15bestFeats.pkl')
```
